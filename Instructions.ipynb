## Graphs_and_topology_for_chemistry

This module includes code to load in PDB or mol2 coordinate files, create molecular structures (courtesy of rdkit), create topological data analysis features (courtesy of [[[giano]]]), output the features as both `.csv` and `.hdf5` files. Examples of plotting structures and graphs and persistence diagrams are included.

You will want to create and save the features rather than create them fresh each time as they take a little while to create. Note that the creation of TDAF is not completely deterministic, numbers change slightly, likely due to differences in rdkit's creation of the structures.

G&t4chem uses the up to date deepchem and tf2, which has more models and featurisations.

**Warning** currently some models in deepchem 2.6 do not give the benchmark results (tehy give worse results) so you will want to test and compare to results from the 2.3 version of deepchem until the DeepChem team fix this issue.

To try out TDAF on a few molecules, try running `Test_xx_first.ipynb`, you will want to start here if you're playing with a new dataset. 

To create the TDAF datasets, use `Do_xx` or `Create`, these will need to be slightly rewritten depending on the dataset (you will have different data to add in addition to the TDAF dataset). 

Note that the refined dataset .csv is the full dataset, the hdf5 dataset will exclude any complexes in the core dataset (based on PDBcode) and you will need to add how many of these there are (as `missing`)

Once you have the features in an hdf5 file, you can use any ML program to train on them. To benchmark the datasets and compare to the deepchem paper, we suggest you use the deepchem_2.3_validation software here [[]]].
