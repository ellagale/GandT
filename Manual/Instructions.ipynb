# Instructions

## To use PHF features

The datasets as used in the paper are in `datasets/after_processing` and can be used immediately. The `.csv` files only work for datasets small enough to be read into memory and are split into $x$ and $y$ data with x being the input features and y being the output target(s). The `.hdf5` are best for large datasets, quicker to use, contain the extra features and more robust. Alternatively, the `.hdf5` files can be remade by running the `Do_datasetname.py` scripts found in `Create_PHF_scripts`.

## To make PHF features from included datasets

This example is for the Delaney dataset. 

First run `Create_PHF_Scripts/Do_delaney.py`. It is strongly suggested that you set the flag `testing` to `True` which will make a very small dataset when first testing the script. The script featurises each molecule in turn and saves out the persistent homology features in `output/converted/x_data_delaney.csv` and the target in `output/converted/y_data_delaney.csv`. the `x_data` features are 18 numbers long and are in the order: Persistence entropy, number of points, bottleneck, wasserstein, landscape, persistence_image and each of these features are three numbers as we are using 3-D space. Note that, some molecules may not be able to be functionalised due to dirty structures, you can either clean them or skip them.

You can directly use the .csv files in your machine learning program. Or, the second half of the Do_delaney script reads in the just created x_data and y_data files, combines that data with the extra features in `datasets/delaney` and optionally creates principal component analysis (PCA) features from the PHF features. 

The scripts `Create_PHF_Scripts/rdkit_delaney.py` will use the structures given and the `rdkit` library to calculate and save out a large number (208) chemical features. 

Note that, if you wish to combine these datasets or selected subsets of features of these datasets (as done in Experiment 3 in the paper) you will want to use the experimentation project on GitHub.



## How to featurise your own datasets

See the Delaney dataset and scripts for a quick to run example.

First create `.csv` files for each dimension of the non-structural input data. If you have very many dimensions of inputs, you may want to script this to create these files automatically and place them in `datasets/dataset_name`. 

Chemical structures can be input as SMILES strings, `.mol`, `.mol2` files or `.pdb` files. If SMILES strings, create a 1-D csv file. See the PDBBind datasets for an example of how to structure `.pdb` and `.mol2` file input (files are put in subfolders.

Next make `Do_datasetname.py` file, using one of the given examples as a basis. You will want to change the code for reading in the non-PHF features. 

You can play around with the dataset in the `.ipynb` and test that the `.hdf5` file works in Testing_dataset.ipynb.

## Graphs_and_topology_for_chemistry

This module includes code to load in PDB or mol2 coordinate files, create molecular structures (courtesy of rdkit), create topological data analysis features (courtesy of [[[giano]]]), output the features as both `.csv` and `.hdf5` files. Examples of plotting structures and graphs and persistence diagrams are included.

You will want to create and save the features rather than create them fresh each time as they take a little while to create. Note that the creation of TDAF is not completely deterministic, numbers change slightly, likely due to differences in rdkit's creation of the structures.

G&t4chem uses the up to date deepchem and tf2, which has more models and featurisations.

**Warning** currently some models in deepchem 2.6 do not give the benchmark results (tehy give worse results) so you will want to test and compare to results from the 2.3 version of deepchem until the DeepChem team fix this issue.

To try out TDAF on a few molecules, try running `Test_xx_first.ipynb`, you will want to start here if you're playing with a new dataset. 

To create the TDAF datasets, use `Do_xx` or `Create`, these will need to be slightly rewritten depending on the dataset (you will have different data to add in addition to the TDAF dataset). 

Note that the refined dataset .csv is the full dataset, the hdf5 dataset will exclude any complexes in the core dataset (based on PDBcode) and you will need to add how many of these there are (as `missing`)




N.B. the file `Do_PDBBind_topology_features_2015_for_ML.py` **does not remove the CORE dataset from the REFINED dataset.** You must run `fix_PDBBind_2015.ipynb` on if afterwards. This problem does not apply to the other PDBBind datasets (2016 onwards).
