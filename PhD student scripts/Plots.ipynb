```python
all_models = pd.read_csv("./models/all_models.csv")
```

```python
df = pd.DataFrame((all_models),
    columns=['PMC', 'GC', 'RMC', 'W', 'MTC'])
```

```python
df.columns
```

```python
my_data = []
for i in [x for x in df.columns]:
    my_data.append(df[i].to_list())
```

```python
my_data
```

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np

# Creating dataset
#np.random.seed(10)
#data_1 = np.random.normal(100, 10, 200)
#data_2 = np.random.normal(90, 20, 200)
#data_3 = np.random.normal(80, 30, 200)
#data_4 = np.random.normal(70, 40, 200)
#data = [data_1, data_2, data_3, data_4]

fig = plt.figure(figsize =(7, 5))
ax = fig.add_subplot(111)

# Creating axes instance
bp = ax.boxplot(my_data, patch_artist = True,
				vert = 1)

colors = ['#4F94CD', '#FF7D40',
		'#CD2626', '#008B00', '#FFC125']

for patch, color in zip(bp['boxes'], colors):
	patch.set_facecolor(color)

# changing color and linewidth of
# whiskers
for whisker in bp['whiskers']:
	whisker.set(color ='#000000',
				linewidth = 2,
				linestyle =":")

# changing color and linewidth of
# caps
for cap in bp['caps']:
	cap.set(color ='#000000',
			linewidth = 2)

# changing color and linewidth of
# medians
for median in bp['medians']:
	median.set(color ='black',
			linewidth = 2)

# changing style of fliers
for flier in bp['fliers']:
	flier.set(marker ='D',
			color ='#000000',
			alpha = 0.5)
    
#changing box outline thickness
for box in bp['boxes']:
    box.set(linewidth = 2)

# x-axis labels
ax.set_xticklabels(['PMC', 'GC',
					'RMC', 'W', 'MTC'])


plt.yticks(fontsize = 20)
plt.xticks(fontsize = 20)

ax.patch.set_edgecolor('black') 

# Removing top axes and right axes
# ticks
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()

ax.set_ylabel('ROC-AUC')
ax.set_xlabel('Models')



#ax.spines['bottom'].set_color('k')
#ax.spines['top'].set_color('k')
#ax.spines['right'].set_color('k')
#ax.spines['left'].set_color('k')
	
# show plot
plt.show(bp)
```
