# Script to remove core datapoints from the 2015 refined dataset

DeepChem / MoleculeNet offers the full refined dataset **which includes most of the clusters in the core dataset**.

Whilst this makes sense for testing scoring functions, it is not the way a machine learning programmer would expect. 

So, run the code in `do_PDBBind_2015.py` to create the datasets and then run this code to remove the extra values. 

N.B. if you get RMSE values and $R^2$ values of 0.8, check that the refined dataset does not include the core values. 

**Trust me on this!**

```python
import h5py
import numpy as np
```

```python
core = h5py.File("F:/Nextcloud/science/Datasets/topol_datasets/PDBBind_core_topological_features.hdf5",'r')
topo = h5py.File("F:/Nextcloud/science/Datasets/topol_datasets/PDBBind_refined_2015_topological_features.hdf5",'r+')
```

```python
print(len(list(core['PDB_code'])))
print(len(list(topo['PDB_code'])))
core_codes = np.array(core['PDB_code'])
topo_codes = np.array(topo['PDB_code'])
```

```python
keep_list=~np.isin(topo_codes,core_codes)
np.count_nonzero(keep_list)
```

```python
print(topo['PDB_code'][0:3])
print(topo['PDB_code'][keep_list][0:3])
print('3i3b' in core_codes)
```

```python
for key in topo:
    old=np.array(topo[key])
    del topo[key]
    topo.create_dataset(key,data=old[keep_list])    
```

```python
core.close()
topo.close()
```
